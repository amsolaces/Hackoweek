```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from sklearn.linear_model import LinearRegression
```

```python
hours = pd.date_range(start="2026-03-01", periods=168, freq="H")  # 1 week hourly

usage = []

for h in hours:
    if 18 <= h.hour <= 22:
        usage.append(np.random.randint(250,350))  # evening peak
    else:
        usage.append(np.random.randint(100,180))

df = pd.DataFrame({
    "timestamp": hours,
    "electricity_kwh": usage
})

df.head()
```

```python
df['moving_avg'] = df['electricity_kwh'].rolling(window=3).mean()

df.head()
```

```python
df['hour_index'] = np.arange(len(df))

X = df[['hour_index']]
y = df['electricity_kwh']

model = LinearRegression()
model.fit(X,y)

df['predicted'] = model.predict(X)

df.head()
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df['timestamp'],
    y=df['electricity_kwh'],
    mode='lines',
    name='Actual Usage'
))

fig.add_trace(go.Scatter(
    x=df['timestamp'],
    y=df['moving_avg'],
    mode='lines',
    name='Moving Average'
))

fig.add_trace(go.Scatter(
    x=df['timestamp'],
    y=df['predicted'],
    mode='lines',
    name='Predicted Usage'
))

fig.show()
```

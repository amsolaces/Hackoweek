# ❄️ Smart HVAC Optimization System for Laboratory Buildings

This project predicts cooling requirements in laboratory spaces using occupancy and temperature data.

A Decision Tree model is trained to forecast HVAC cooling demand, and predictions are visualized through a zone-wise heatmap dashboard.

Features:
• Occupancy-aware HVAC control  
• Temperature-based cooling prediction  
• Decision Tree machine learning model  
• Zone heatmap visualization

```python
import pandas as pd
import numpy as np
import plotly.express as px
from sklearn.tree import DecisionTreeRegressor
```

```python
zones = ["Lab A","Lab B","Lab C","Lab D","Lab E"]

data = []

for i in range(200):

    zone = np.random.choice(zones)
    occupancy = np.random.randint(0,30)
    temperature = np.random.randint(22,35)

    cooling_need = occupancy*2 + temperature*3 + np.random.randint(5,20)

    data.append([zone,occupancy,temperature,cooling_need])

df = pd.DataFrame(data, columns=[
    "zone",
    "occupancy",
    "temperature",
    "cooling_demand"
])

df.head()
```

```python
X = df[['occupancy','temperature']]
y = df['cooling_demand']

model = DecisionTreeRegressor()
model.fit(X,y)

df["predicted_cooling"] = model.predict(X)
```

```python
zone_demand = df.groupby("zone")["predicted_cooling"].mean().reset_index()
```

```python
heatmap_data = df.pivot_table(
    values="predicted_cooling",
    index="zone",
    columns="occupancy",
    aggfunc="mean"
)

fig = px.imshow(
    heatmap_data,
    labels=dict(color="Cooling Demand"),
    title="Lab HVAC Cooling Demand Heatmap"
)

fig.show()
```

```python
threshold = 120

high_demand = df[df["predicted_cooling"] > threshold]

print("Zones requiring high cooling demand:")
print(high_demand["zone"].unique())
```

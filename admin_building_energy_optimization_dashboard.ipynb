# 🏢 Admin Building Energy Optimization Dashboard

This project analyzes electricity consumption patterns in an administrative building to detect weekend energy dips and identify potential savings opportunities.

Using K-Means clustering, the system groups energy usage behaviors and forecasts consumption trends.

Features:
• Weekend vs weekday usage analysis  
• K-Means clustering of energy profiles  
• Forecasting of usage clusters  
• Visual dashboard with energy savings insights

```python
import pandas as pd
import numpy as np
import plotly.express as px
from sklearn.cluster import KMeans
from sklearn.linear_model import LinearRegression
```

```python
days = pd.date_range(start="2026-01-01", periods=120)

usage = []
day_type = []

for d in days:

    if d.weekday() < 5:
        usage.append(np.random.randint(350,500))
        day_type.append("Weekday")

    else:
        usage.append(np.random.randint(120,220))
        day_type.append("Weekend")

df = pd.DataFrame({
    "date": days,
    "energy_usage": usage,
    "day_type": day_type
})

df.head()
```

```python
X = df[['energy_usage']]

kmeans = KMeans(n_clusters=3)
df['cluster'] = kmeans.fit_predict(X)
```

```python
df["time_index"] = np.arange(len(df))

X = df[['time_index']]
y = df['energy_usage']

model = LinearRegression()
model.fit(X,y)

df['predicted_usage'] = model.predict(X)
```

```python
fig = px.line(
    df,
    x="date",
    y=["energy_usage","predicted_usage"],
    color="day_type",
    title="Admin Building Energy Usage Trends"
)

fig.show()
```

```python
fig = px.scatter(
    df,
    x="date",
    y="energy_usage",
    color="cluster",
    title="Energy Usage Clustering"
)

fig.show()
```

```python
weekday_avg = df[df["day_type"]=="Weekday"]["energy_usage"].mean()
weekend_avg = df[df["day_type"]=="Weekend"]["energy_usage"].mean()

savings = weekday_avg - weekend_avg

pie_data = pd.DataFrame({
    "Category":["Current Usage","Potential Savings"],
    "Value":[weekend_avg,savings]
})

fig = px.pie(
    pie_data,
    names="Category",
    values="Value",
    title="Weekend Energy Saving Potential"
)

fig.show()
```

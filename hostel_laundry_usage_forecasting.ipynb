# 🧺 Hostel Laundry Usage Forecasting System

This project analyzes laundry machine usage in hostel facilities and predicts peak demand periods.

Naive Bayes is used to classify usage categories and Prophet forecasting predicts future laundry demand patterns.

Features:
• Laundry usage categorization  
• Peak demand prediction  
• Time-series forecasting  
• Interactive timeline visualization

```python
import pandas as pd
import numpy as np
import plotly.express as px
from prophet import Prophet
from sklearn.naive_bayes import GaussianNB
```

```python
days = pd.date_range(start="2026-01-01", periods=120)

usage = []

for d in days:

    if d.weekday() in [5,6]:
        usage.append(np.random.randint(200,350))
    else:
        usage.append(np.random.randint(80,160))

df = pd.DataFrame({
    "ds": days,
    "y": usage
})
```

```python
model = Prophet()

model.fit(df)

future = model.make_future_dataframe(periods=30)

forecast = model.predict(future)
```

```python
fig = px.line(
    forecast,
    x="ds",
    y="yhat",
    title="Hostel Laundry Demand Forecast"
)

fig.show()
```

# 🍽 Smart Cafeteria Load Prediction System

This project predicts cafeteria demand during lunch hours using weather and temperature data.

Using Linear Regression, the system forecasts cafeteria crowd levels and visualizes real-time demand trends through an interactive dashboard.

Features:
- Weather-aware demand prediction
- Lunch-hour surge detection
- Linear regression forecasting
- Real-time updating visualization

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go
from sklearn.linear_model import LinearRegression
```

```python
hours = pd.date_range(start="2026-03-01", periods=200, freq="H")

temperature = np.random.randint(22, 36, size=len(hours))

visitors = []

for h, temp in zip(hours, temperature):

    if 12 <= h.hour <= 14:  # lunch hours
        visitors.append(np.random.randint(120,220) + temp)
    else:
        visitors.append(np.random.randint(20,80))

df = pd.DataFrame({
    "timestamp": hours,
    "temperature": temperature,
    "cafeteria_visitors": visitors
})

df.head()
```

```python
X = df[['temperature']]
y = df['cafeteria_visitors']

model = LinearRegression()
model.fit(X, y)

df['predicted_load'] = model.predict(X)
```

```python
lunch_peak = df[df["timestamp"].dt.hour.between(12,14)]

print("Average Lunch Demand:", lunch_peak["cafeteria_visitors"].mean())
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=df["timestamp"],
    y=df["cafeteria_visitors"],
    mode="lines",
    name="Actual Demand"
))

fig.add_trace(go.Scatter(
    x=df["timestamp"],
    y=df["predicted_load"],
    mode="lines",
    name="Predicted Demand"
))

fig.update_layout(
    title="Smart Cafeteria Demand Prediction Dashboard",
    xaxis_title="Time",
    yaxis_title="Cafeteria Visitors"
)

fig.show()
```

```python
import time
from IPython.display import clear_output

for i in range(5):

    new_temp = np.random.randint(22,36)
    predicted_visitors = model.predict([[new_temp]])[0]

    print("Current Temperature:", new_temp)
    print("Predicted Cafeteria Demand:", int(predicted_visitors))

    time.sleep(2)
    clear_output(wait=True)
```
